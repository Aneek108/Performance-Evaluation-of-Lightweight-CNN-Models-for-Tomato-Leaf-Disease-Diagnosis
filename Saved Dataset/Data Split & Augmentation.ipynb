# Random Samples

```python
import os
import cv2
import random
import matplotlib.pyplot as plt
```

```python
data_dir = 'E:/Research & Projects/Tomato Disease Detection/Saved Dataset/Original'
class_names = sorted(os.listdir(data_dir))

plt.figure(figsize=(15, 6))
for i, class_name in enumerate(class_names):

    class_path = f"{data_dir}/{class_name}"
    image_name = random.sample(os.listdir(class_path), 1)[0]
    image_path = f"{class_path}/{image_name}"

    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    plt.subplot(2, 5, i + 1)
    plt.imshow(image)
    plt.title(class_name)

plt.tight_layout()
plt.show()
```
